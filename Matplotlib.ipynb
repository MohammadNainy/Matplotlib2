```python
fig , (ax1,ax2) = plt.subplots(2)
ax1.hist(df["FUELCONSUMPTION_COMB"],range=(0,30),histtype='step');
ax1.set(title="Graph1",
        xlabel="Fuel",
        ylabel="Number"
       );
ax2.boxplot(df["FUELCONSUMPTION_COMB"],labels=["ALL"],showfliers=False,vert=False);
ax2.set(title="Graph",
       xlabel="Fuel");
```

```python
fig , (ax1,ax2) = plt.subplots(2)
ax1.hist(df["FUELCONSUMPTION_COMB"],range=(0,30),histtype='step');
ax1.set(title="Graph1",
        xlabel="Fuel",
        ylabel="Number"
       );
ax2.boxplot(df["FUELCONSUMPTION_COMB"],labels=["ALL"],showfliers=False,vert=False);
ax2.set(title="Graph",
       xlabel="Fuel");

fig.suptitle("Total",fontsize="16",fontweight="bold")
```

```python
fig , (ax1,ax2) = plt.subplots(1,2)
ax1.hist(df["FUELCONSUMPTION_COMB"],range=(0,30));
ax1.set(title="Graph1",
        xlabel="Fuel",
        ylabel="Number"
       );
ax2.boxplot(df["FUELCONSUMPTION_COMB"],labels=["ALL"],showfliers=False);
ax2.set(title="Graph",
       ylabel="Fuel");

fig.suptitle("Total",fontsize="16",fontweight="bold")
```

```python
fig , (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(10,4) )
ax1.hist(df["FUELCONSUMPTION_COMB"],range=(0,30));
ax1.set(title="Graph1",
        xlabel="Fuel",
        ylabel="Number"
       );
ax2.boxplot(df["FUELCONSUMPTION_COMB"],labels=["ALL"],showfliers=False);
ax2.set(title="Graph",
       ylabel="Fuel");

fig.suptitle("Total",fontsize="16",fontweight="bold")
```
