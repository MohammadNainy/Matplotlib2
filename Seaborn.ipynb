```python
crt = pd.crosstab(df["CYLINDERS"],
                  df["FUELTYPE"],
                  values=df["FUELCONSUMPTION_COMB_MPG"],
                  aggfunc="mean"
                 )
crt
```

```python
sns.heatmap(crt)
```

```python
sns.heatmap(crt,
           annot=True,
           fmt=".0f", 
           cmap="mako" 
           );
```

```python
fig , ax = plt.subplots(figsize=(9,6))
sns.heatmap(crt,
           annot=True,
           fmt=".0f",
           cmap="mako",
           ax=ax
           );
ax.set_ylabel("Cylin")
fig.savefig("seaborn_heatmap.svg")
```
